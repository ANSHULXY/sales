```python
file['Date']= pd.to_datetime(file['Date'],errors='coerce')
```

```python
file[file['Date'].isnull()]
```

```python
#  Filling the NaT  values with average of time
```

```python
file["Date"].fillna(file["Date"].mean(),inplace=True)
```

```python
file[file['Date'].isnull()]
```

```python
file[file['Unnamed: 0']==1261]
```

```python
#fetching month,day of week, weekday
file["month"]=file["Date"].dt.month_name()
file["day"]=file["Date"].dt.day_name()
file["dayoftheweek"]=file["Date"].dt.weekday
file["year"]=file["Date"].dt.year
```

```python
file.head()
```

```python
## Droping column unnamed as it is not usefull for us
```

```python
file.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
file.sample()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(file.corr(),annot=True)
```

```python
#no two colum has a good relation
```

```python
# Total unit sales Product 1, Product 2, Product 3, Product 4 
q = file[["Q-P1","Q-P2","Q-P3","Q-P4"]].sum()
print(q)
plt.figure(figsize=(8,8))
plt.pie(q,labels=file[["Q-P1","Q-P2","Q-P3","Q-P4"]].sum().index,shadow=True,autopct="%0.01f%%",textprops={"fontsize":20},wedgeprops={'width': 0.8},explode=[0,0,0,0.3])
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.8));
```

```python
# Total Revenue percent from sales from Product 1, Product 2, Product 3, Product 4  
s=file[["S-P1","S-P2","S-P3","S-P4"]].sum()
print(s)
plt.figure(figsize=(8,8))
plt.pie(s,labels=file[["S-P1","S-P2","S-P3","S-P4"]].sum().index,shadow=True,autopct="%0.01f%%",textprops={"fontsize":20},wedgeprops={'width': 0.8},explode=[0,0,0,0.3])
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.8))
```

```python
# sales according to months
print(file["month"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="month",data=file,edgecolor="black")
plt.xticks(rotation=90);
```

```python
# which is the most occuring Day 
print(file["day"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="day",data=file,edgecolor="black")
plt.xticks(rotation=90);
```

```python
# which is the most occuring year 
print(file["year"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="year",data=file,edgecolor="black")
plt.xticks(rotation=90);
```

```python
# Monthly distrubution of revenue 
sns.relplot(x="month",y="S-P1",data=file,kind="line",height=10,color="red")
plt.xticks(rotation=90);
sns.relplot(x="month",y="S-P2",data=file,kind="line",height=10,color="blue")
plt.xticks(rotation=90);
sns.relplot(x="month",y="S-P3",data=file,kind="line",height=10,color="green")
plt.xticks(rotation=90);
sns.relplot(x="month",y="S-P4",data=file,kind="line",height=10,color="purple")
plt.xticks(rotation=90);
```

```python
## In which month revenue was it peak
file.groupby("month")[["S-P1","S-P2","S-P3","S-P4"]].sum()
```

```python
plt.figure(figsize=(15,15),dpi=100)
plt.subplot(2,2,1)
sns.barplot(x="month",y="S-P1",data=file,edgecolor="black",estimator=sum)
plt.xticks(rotation=90);
plt.subplot(2,2,2)
sns.barplot(x="month",y="S-P2",data=file,edgecolor="black",estimator=sum)
plt.xticks(rotation=90);
plt.subplot(2,2,3)
sns.barplot(x="month",y="S-P3",data=file,edgecolor="black",estimator=sum)
plt.xticks(rotation=90);
plt.subplot(2,2,4)
sns.barplot(x="month",y="S-P4",data=file,edgecolor="black",estimator=sum)
plt.xticks(rotation=90)
plt.subplots_adjust(hspace=0.3);
```
